```python
#format string
def num():
    txt="some days makes happy"
    var1="100"
    val_1="of smile"
    value=f"Day of light,{txt} {var1} {val_1}"
    print(value)
num()
```

```python
#decimal
def var():
    milk=10
    butter=5
    chocolate=30
    prod=f"The rate of dairy products is {milk:.3f} {butter:.2f} {chocolate:.1f}dollars"
    print(prod)
var()
```

```python
#string
def name():
    val="cherry"
    fru=f"The red fruit is {val:.2f}dollars"
name()
print(fru)
```

```python
#escape character
def esc():
    txt_4="A trusting little \"Leaf\" of green."
    print(txt_4)
esc()
```

```python
#character
def char():
    text="The \'Plates\' will still shift and the clouds will still spew\nThe sun will slowly rise and the moon will follow too."
    print(text)
char()
```

```python
text="The sun will \b slowly rise"
print(text)
```
